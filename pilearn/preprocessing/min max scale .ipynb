```python
import numpy as np
```

```python
X = np.linspace(-10, 10, 100)
y = [x*x + np.random.randn() for x in X]
X = np.c_[np.ones(len(X)), X]
```

```python
y = np.array(y).reshape(-1, 1)
```

```python
n_samples, n_features = X.shape
```

```python
def min_max_scaler(X):
    list_of_arrays = list()
    for i in X.T:
        if len(set(i)) == n_samples:
            normed_val = [(j - min(i)) / (max(i) - min(i)) for j in i]
            list_of_arrays.append(normed_val)
        else:
            list_of_arrays.append(i)
    return np.c_[list_of_arrays].T
```

```python
def mean_variance_scaler(X):
    list_of_arrays = list()
    for i in X.T:
        if len(set(i)) == n_samples:
            normed_val = [(j - np.mean(i)) / np.std(i) for j in i]
            list_of_arrays.append(normed_val)
        else:
            list_of_arrays.append(i)
    return np.c_[list_of_arrays].T
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(X[:, 1], y)
plt.title('unscaled')
plt.show()
```

```python
plt.plot(min_max_scaler(X)[:,1], min_max_scaler(y))
#plt.plot(mean_variance_scaler(X)[:,1], mean_variance_scaler(y), 'r')
plt.title('min max scale')
plt.show()
```

```python
plt.plot(mean_variance_scaler(X)[:,1], mean_variance_scaler(y))
plt.title('mean var scaler')
plt.show()
```
